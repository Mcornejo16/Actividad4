```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv('listings2.csv')
df
```

```python
df.info()
```

```python
# a) Los registros de 5 “host name” diferentes
filtro_a = df[df["host_name"].isin(df["host_name"].unique()[:5])]
filtro_a
```

```python
filtro_a.to_csv("Part_A.csv")
```

```python
# b) Los registros de los host que se hayan unido a Airbnb después del año 2020
# Convertimos la columna a datetime si no lo está
df["host_since"] = pd.to_datetime(df["host_since"], errors="coerce")
filtro_b = df[df["host_since"].dt.year > 2020]
filtro_b
```

```python
filtro_b.to_csv("Part_B.csv")
```

```python
# c) Los registros de los host que responden “a más tardar en 1 día”
filtro_c = df[df["host_response_time"] == "within a day"]
filtro_c
```

```python
filtro_c.to_csv("Part_C.csv")
```

```python
# d) Los registros de las filas múltiplos de 200 (400, 600, 800, etc.)
filtro_d = df.iloc[[i for i in range(len(df)) if i % 200 == 0 and i != 0], :]
filtro_d
```

```python
filtro_d.to_csv("Part_D.csv")
```

```python
filtro_e = df[df["room_type"].isin(["Entire home/apt", "Private room"])]
filtro_e
```

```python
filtro_e.to_csv("Part_E.csv")
```

```python
# f) Los registros que “a lo mucho cuenten con 4 camas”
filtro_f = df[df["beds"] <= 4]
filtro_f
```

```python
filtro_f.to_csv("Part_F.csv")
```

```python
# g) Los registros que “sean superhost”
filtro_g = df[df["host_is_superhost"] == "t"]
filtro_g
```

```python
filtro_g.to_csv("Part_G.csv")
```

```python
# h) Los registros de las columnas impares
filtro_h = df.iloc[:, 1::2]  # Selecciona columnas con índices 1,3,5...
filtro_h
```

```python
filtro_h.to_csv("Part_H.csv")
```
